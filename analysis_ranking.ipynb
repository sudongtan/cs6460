```python
%load_ext autoreload
%autoreload 2
from recommend import load_model, recommend
import warnings
warnings.filterwarnings('ignore')
from collections import defaultdict
import pandas as pd
import numpy as np
```

```python
try:
    model = load_model()
except ValueError:
    pass
```

```python
sample_size = 10
np.random.seed(333)

nolikes = defaultdict(lambda: {'random': 0, 'interested_only': 0, 'topic': 0})
scores = defaultdict(lambda: {'random': 0, 'interested_only': 0, 'topic': 0})
interested = defaultdict(lambda: {'random': 0, 'interested_only': 0, 'topic': 0})
relevances = defaultdict(lambda: {'random': 0, 'interested_only': 0, 'topic': 0})

user_ids = np.random.choice(range(45963), sample_size, replace=False)
for i, user_id in enumerate(user_ids):
    if i % 100 == 0:
        print(i, user_id)
    for method in ['random', 'interested_only', 'topic']:
        liked, notliked, i, score, relevance = recommend(model, user_id, k=10, mode=method, display=False)
        nolikes[user_id][method] = len(notliked)
        scores[user_id][method] = score
        interested[user_id][method] = i
        relevances[user_id][method] = relevance
```

```python
df = pd.DataFrame.from_dict(nolikes, 'index')
df.to_csv('results_demo/ranking_nolikes.csv', index=False)
df = pd.DataFrame.from_dict(scores, 'index')
df.to_csv('results_demo/ranking_scores.csv', index=False)
df = pd.DataFrame.from_dict(relevances, 'index')
df.to_csv('results_demo/ranking_relevances.csv', index=False)
df = pd.DataFrame.from_dict(interested, 'index')
df.to_csv('results_demo/ranking_interested.csv', index=False)
```
